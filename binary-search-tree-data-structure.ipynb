```python
class tree():
    def __init__(self,val=None):
        self.root=val
        self.left= None
        self.right=None

    
    # Print Pre-Order
    def preorder_traversal(self):
        if self.root is not None:
            print(self.root, end=" ")
        if self.left is not None:
            self.left.preorder_traversal()
        if self.right is not None:
            self.right.preorder_traversal()    

    #Print Post-Order
    def postorder_traversal(self):
        if self.left is not None:
            self.left.postorder_traversal()
        if self.right is not None:
            self.right.postorder_traversal()
        print (self.root,end=" ")
        
    #Print Inorder
    def inorder_traversal(self):
        if self.left is not None:
            self.left.inorder_traversal()
        print (self.root, end=" ")
        if self.right is not None:
            self.right.inorder_traversal()    

    #Determine the level of the tree
    def depth(self):
        if self is None:
            return 0
        else:
            if(self.left is not None):
                lh=self.left.depth()
            else:
                lh=0
            if(self.right is not None):
                rh=self.right.depth()      
            else:
                rh=0      
            return max(lh+1,rh+1)            
    
    #Determine the values at each level in the tree
    def level_node(self,dt,leaf_arr):
        if self is None:
            return 
        if dt ==1 :
            leaf_arr.append(self.root)
        elif (dt>1):
            if self.left is not None:
                self.left.level_node(dt-1,leaf_arr)
            if self.right is not None:
                self.right.level_node(dt-1, leaf_arr)  
        return
    
    def leaf_max_min_all(self):
        total_depth=self.depth()
        for d in range(1,total_depth+1):
            leaf_arr=[]
            self.level_node(d,leaf_arr)
            print(f"level: {d} ELements: {leaf_arr}")
            
            
    #Check if the value matches with the current node. If node, check if the value is less than the current node or more. In case the value is less, 
    #repeat the search in the left node else peform the same operations in the righ node 
    def find_element(self,val):
        if self is None:
            return -1
        elif (self.root == val):
            return 1
        elif (val < self.root):
            if self.left is not None:
                return (self.left.find_element(val))
            else:
                return -1
        else:
            if self.right is not None:
                return (self.right.find_element(val))
            else:
                return -1
    
    #find the minimum value in the right node:
    def find_min_value(self):
        if self.left is None:
            return self.root
        if self.left is not None:
            value=self.left.find_min_value()
            return value
        
    #Remove an element from the tree
    def remove_element(self,val):
        #check if the element is less than the root. If yes, look for the left node else right
        if (val < self.root):
            if self.left is not None:
                self.left=self.left.remove_element(val)
                return self
            else:
                return None
        elif (val > self.root):
            if self.right is not None:
                self.right=self.right.remove_element(val)
                return self
            else:
                return None
        else:
            #This is the case when the node has both the left and the right child
            if (self.left is not None and self.right is not None):
                new_node= self.right.find_min_value()
                self.root=new_node
                self.right=self.right.remove_element(new_node)
                return self
            #this is the case when the node has no right or left child
            elif(self.left is None and self.right is None):
                return None
            #This is the case when the node has only left child
            elif (self.left is not None):
                return self.left
            #This is the case when the node has only right child
            elif (self.right is not None):
                return self.right
            return self    
        
    #Add an element to the tree. If the value is greater to the root, traverse to the right node else traverse to the left node
    #once the position is determined,create a new node and add to the existing node
    def add_element(self,val):
        if self is None:
            return None
        elif (val< self.root):
            #Traverse to the left
            if (self.left is not None):
                self.left=self.left.add_element(val)
            else:
                self.left=tree(val)
        elif(val>self.root):
            #Traverse to the right
            if(self.right is not None):
                self.right=self.right.add_element(val)
            else:
                self.right=tree(val)
            
        return self
    
    #Print the left side of the tree
    def left_side_view(self,larr):
        if self is not None:
            larr.append(self.root)
        else:
            return
        if (self.left is not None):
            self.left.left_side_view(larr)
        return    
    
    #print the right side of the tree
    def right_side_view(self,rarr):
        if self is not None:
            rarr.append(self.root)
        else:
            return
        if self.right is not None:
            self.right.right_side_view(rarr)
        return 
    
    #print all non-leaves of the tree
    def print_non_leave(self,result):
        if self is None:
            return
        elif (self.left is None and self.right is None):
            return 
        else:
            result.append(self.root)
        if self.left is not None:
            self.left.print_non_leave(result)
        if self.right is not None:
            self.right.print_non_leave(result)
        return 
    
    #print all leaves of the tree
    def print_leave(self,result):
        if self is None:
            return
        elif (self.left is None and self.right is None):
            result.append(self.root)
        if self.left is not None:
            self.left.print_leave(result)
        if self.right is not None:
            self.right.print_leave(result)
            
        return     
```

```python
def create_tree(arr,tree):
    t=tree(arr[0])
    for i in range(1,len(arr)):
        t.add_element(arr[i])
    return t
```

```python
#Create the tree 
arr=[22, 12, 30, 8, 20, 25,40]
t=create_tree(arr,tree)
```

```python
print("Pre-order")
t.preorder_traversal()
print("\nPost-order")
t.postorder_traversal()
print("\nIn-order")
t.inorder_traversal()
print("\nTotal Depth of the Tree")
t.depth()
print("\nBreadth Search- Data at each level")
t.leaf_max_min_all()
print("\nSearch for an element")
t.find_element(250)
print("\nWith new element in the tree")
t.add_element(13)
t.preorder_traversal()
print("\nRemove element in the tree")
t.remove_element(13)
t.preorder_traversal()
print("\nLeft view of the tree")
larr=[]
t.left_side_view(larr)
print(larr)
print("\nRight view of the tree")
rarr=[]
t.right_side_view(rarr)
print(rarr)
print("\nNon Leaves in the tree")
non_leave_arr=[]
t.print_non_leave(non_leave_arr)
print(non_leave_arr)
print("\nLeaves in the tree")
leave_arr=[]
t.print_leave(leave_arr)
print(leave_arr)
```
